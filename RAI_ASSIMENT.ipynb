**SHAP**

This gives a simple example of explaining a linear logistic regression sentiment analysis model using shap. Note that with a linear model, the SHAP value of feature
 for the prediction
 (assuming feature independence) is just
. Since we are explaining a logistic regression model, the units of the SHAP values will be in the log-odds space.

The dataset we are using is the classic IMDB dataset from this paper. When explaining the model, it is interesting to observe how the words that are absent from the text are sometimes just as important as those that are present.

```python
import numpy as np
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

import shap

np.random.seed(101)
shap.initjs()
```

```python
# Import necessary libraries
import shap
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Load the IMDB dataset
corpus, y = shap.datasets.imdb()

# Split the dataset into train and test sets
corpus_train, corpus_test, y_train, y_test = train_test_split(corpus, y, test_size=0.2, random_state=7)

# Create a TfidfVectorizer to transform the text data into numerical features
vectorizer = TfidfVectorizer(min_df=10)
X_train = vectorizer.fit_transform(corpus_train).toarray()  # Convert to array for further use
X_test = vectorizer.transform(corpus_test).toarray()  # Use the same transformer for test data

# Check the shapes of the resulting feature matrices
print(X_train.shape, X_test.shape)
```

```python
# FIT a linear logistic regression model
model = sklearn.linear_model.LogisticRegression(penalty="l2", C=0.1)
model.fit(X_train, y_train)
print(classification_report(y_test, model.predict(X_test)))
```

```python
#Explain the linear model
explainer = shap.Explainer(model, X_train, feature_names=vectorizer.get_feature_names_out())
shap_values = explainer(X_test)
```

```python
shap.plots.beeswarm(shap_values)
```

```python
#Explain the first review’s sentiment prediction
import shap
shap.initjs()
ind = 0
shap.plots.force(shap_values[ind])
```

```python
print("Positive" if y_test[ind] else "Negative", "Review:")
print(corpus_test[ind])
```

```python
#Explain the second review’s sentiment prediction
import shap
shap.initjs()
ind = 1
shap.plots.force(shap_values[ind])
```

```python
print("Positive" if y_test[ind] else "Negative", "Review:")
print(corpus_test[ind])
```

```python
#Explain the third review’s sentiment prediction
import shap
shap.initjs()
ind = 2
shap.plots.force(shap_values[ind])
```

```python
print("Positive" if y_test[ind] else "Negative", "Review:")
print(corpus_test[ind])
```
